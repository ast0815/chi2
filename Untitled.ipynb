```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import chi2
```

```python
cov = np.array([[5,4,3,2,1], [4,5,4,3,2], [3,4,5,4,3], [2,3,4,5,4], [1,2,3,4,5]])
cov
```

```python
rng = np.random.default_rng()
mu = [ 10, 11, 12, 13, 14]
data = rng.multivariate_normal(mu, cov, size=10000)
data
```

```python
d = (data - mu)[:,np.newaxis,:] @ np.linalg.inv(cov) @ (data - mu)[:,:,np.newaxis]
d = d.flatten()
```

```python
plt.hist(d, density=True, bins=20, label="Data")
x = np.linspace(0, 20, 50)
y = chi2.pdf(x, df=5)
plt.plot(x, y, label="$\chi^2_5$")
plt.legend()
```
